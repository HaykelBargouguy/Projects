```python
class Name_company_languge_detection:
    def __init__(self):
        self.nlp = spacy.blank('en')
        self.nlp.add_pipe('sentencizer')  # ajout du sentencizer
        @Language.factory('language_detector')
        def create_language_detector(nlp, name):
            return LanguageDetector()
        self.nlp.add_pipe('language_detector', last=True)

    def detect_language(self, text):
        doc = self.nlp(text)
        return doc._.language['language']
```

```python
lang_detector = Name_company_languge_detection()
```

```python
df['Language'] = df['Name_company'].apply(lambda x: lang_detector.detect_language(x) if isinstance(x, str) else 'Unknown')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.esky.fr/guide-de-voyage/volsbillets-davion/informations-douanesvisas/liste-des-langues-officielles-dans-le-monde-entier'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')
rows = table.find('tbody').find_all('tr')

data = []
for row in rows:
    cols = row.find_all('td')
    if len(cols) >= 2:
        country = cols[0].text.strip()
        language = cols[1].text.strip()
        data.append((country, language))
df_lang = pd.DataFrame(data, columns=['Country', 'Language'])
df_lang
```

# **langdetect**

```python
from langdetect import detect

def detect_lang_langdetect(text):
    try:
        return detect(text)
    except:
        return 'unknown'

df['langdetect'] = df['Name_company'].apply(detect_lang_langdetect)
df.to_csv('name_language_langdetect.csv', index=False)
```

## **langid**

```python
  import langid

def detect_lang_langid(text):
    try:
        lang, _ = langid.classify(text)
        return lang
    except:
        return 'unknown'

df['langid'] = df['Name_company'].apply(detect_lang_langid)
df.to_csv('name_language_langid.csv', index=False)
```
